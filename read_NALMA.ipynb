```python
import re
import pandas as pd
import concurrent.futures
from concurrent.futures import ProcessPoolExecutor,wait,as_completed
```

```python

fname = "NALMA_220712_235000_0600.dat"
expression ='(\d*[.?]\d*\s*)([+-?]\d*[.?]\d*\s*)([+-?]\d*[.?]\d*\s*)([+-?]\d*[.?]\d*\s*)([+-?]\d*[.?]\d*\s*)([+-?]\d*[.?]\d*\s*)([0-9a-zA-Z]*)'
def extract_NALMA(expression,fname):
    with open(fname,mode='rt') as f:
        nalma_data = f.read()
        x = re.findall(expression, nalma_data)
        df = pd.DataFrame(x,columns=['Time (UT sec of day)','Latitude','Longitude', 'Altitude (m)', 'reduced chi^2', 'P(dBW)','mask'])
        df = df.astype({'Time (UT sec of day)':'float32',
                        'Latitude':'float32',
                        'Longitude':'float32',
                        'Altitude (m)':'float32',
                        'reduced chi^2':'float32',
                        'P(dBW)':'float32',
                        'mask':'object'});
    return df
```

```python
df = extract_NALMA(expression,fname)
```
